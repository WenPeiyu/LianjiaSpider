```python
path = r"C:\Users\wenpeiyu\Desktop\rumors_v170613.json"
import pandas as pd 
import thulac
import json

with open(path, "r", encoding="utf-8") as f:
    str_rumor_o = f.readlines()
str_rumor = ""
lst_rumor = []
for num, i in enumerate(str_rumor_o):
    lst_rumor.append(json.loads(i))
df_rumor =  pd.DataFrame(lst_rumor)
str_rumor = " ".join(list(df_rumor['rumorText']))
with open("d:/1.txt", "w", encoding="utf-8") as wf:
    wf.write(str_rumor)
```

```python
thu1 = thulac.thulac(user_dict=None, model_path=None, T2S=False, seg_only=True, filt=False, deli='_')  
text = thu1.cut(str_rumor, text=True) 
```
